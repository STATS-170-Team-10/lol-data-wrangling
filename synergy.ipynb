```python
import pandas as pd
import numpy as np
import itertools
```

```python
names = pd.read_csv('data/game_names.csv').drop(['t1_win'], axis=1)
synergy = pd.read_csv('data/synergy.csv')
```

```python
sdict = {}

for index, row in synergy.iterrows():
    key1 = row[]
    key = row['champId1'] + '-' + row['champName2']
    value = row['synergy_score']
    sdict[key] = value
```

```python
sdict
```

```python
t1_synergy = []
t2_synergy = []

for index, row in names.iterrows():
    ally = list(row)[:5]
    oppo = list(row)[5:]
    t1 = 0
    t2 = 0
    
    for pair in itertools.combinations(ally, 2):
        npair = pair[0] + '-' + pair[1]

        if npair in sdict:
            t1 += sdict[npair]
            
        else:
            npair = pair[1] + '-' + pair[0]
            t1 += sdict[npair]
            
    
    for pair in itertools.combinations(oppo, 2):
        npair = pair[0] + '-' + pair[1]

        if npair in sdict:
            t2 += sdict[npair]
            if index == 97:
                print(npair)
        else:
            npair = pair[1] + '-' + pair[0]
            t2 += sdict[npair]
            if index == 97:
                print(npair)
    
    t1_synergy.append(t1)
    t2_synergy.append(t2)
```

```python
table = [t1_synergy, t2_synergy]
```

```python
df = pd.DataFrame({'t1_synergy': t1_synergy, 't2_synergy': t2_synergy})
```

```python
df.to_csv('data/team_synergy.csv', index=False)
```
